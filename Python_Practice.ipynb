```python
#You are monitoring an electrical substation. The target grid frequency must remain stable at exactly 50.0 Hz. You receive two real-time frequency deviations from separate meters: val_1 = 49.85 and val_2 = 0.15.
#Write a function is_grid_stable(val_1, val_2, target=50.0) that adds val_1 and val_2 together and verifies if they equal the target.


import math
def is_grid_stable(val_1,val_2,target = 50.0):
  return math.isclose(val_1+val_2,target)

print(is_grid_stable(49.85,0.15))
```

```python
"""Write a function called clean_grid_integers(data_list) that takes a list and returns a new list containing only pure integers. It must completely drop any boolean flags.

The Rules:
If an item is a pure integer (like 500 or 0), keep it.

If an item is a boolean flag (True or False), ignore it completely.

The final returned list must be sorted in ascending order (smallest to largest)."""

def clean_grid_integers(data_list):
  return sorted([item for item in data_list if type(item) is int ])
corrupted_data = [120, True, 0, 450, False, 85, True]
print(clean_grid_integers(corrupted_data))
```
