```python
"""In a balanced 3-phase electrical system, the current flowing through all three phases should be roughly equal. If one phase carries significantly more or less current than the others, it indicates an imbalance that can overheat local transformers.Your task is to calculate the absolute difference between the maximum current recorded and the minimum current recorded among the three phases, and check if it exceeds a safe threshold.Your TaskWrite the function logic for check_phase_imbalance(currents_list, max_allowed_imbalance) that:Receives a list of three floats called currents_list (e.g., [150.5, 152.0, 148.5]).Receives a float max_allowed_imbalance.Finds the highest current in the list using Python's built-in max() function, and the lowest current using min().Calculates the imbalance: $\text{imbalance} = \max(\text{currents}) - \min(\text{currents})$.If that calculated imbalance is strictly greater than max_allowed_imbalance, return the string "UNBALANCED_LOAD".Otherwise, return the string "BALANCED_LOAD"."""
def check_phase_imbalance(currents_list, max_allowed_imbalance):
    # --- WRITE YOUR SOLUTION BELOW (Test in Colab first!) ---
    max_load = max(currents_list)
    min_load = min(currents_list)
    diff = abs(max_load - min_load )
    if diff <= max_allowed_imbalance:
      return "BALANCED_LOAD"
    return "UNBALANCED_LOAD"
    # --- WRITE YOUR SOLUTION ABOVE ---


# ==========================================
# 🧪 TEST CASES
# ==========================================

# Test A: Max is 152.0, Min is 149.0. Diff is 3.0 (<= 5.0) -> "BALANCED_LOAD"
print("Test A:", check_phase_imbalance([150.0, 152.0, 149.0], 5.0))

# Test B: Max is 161.0, Min is 148.0. Diff is 13.0 (> 5.0) -> "UNBALANCED_LOAD"
print("Test B:", check_phase_imbalance([148.0, 155.0, 161.0], 5.0))
```

```python
"""The ScenarioIn power systems engineering, alternating current (AC) grids must maintain a precise frequency—exactly $50.0\text{ Hz}$ in India. If the frequency spikes too high due to a sudden drop in customer demand, it can trigger widespread equipment tripping.Your monitoring script receives a history of continuous frequency readings as a list of floats. An "Over-Frequency Event" occurs only if the frequency remains strictly greater than $50.5\text{ Hz}$ for two consecutive readings.Your TaskWrite the function logic for detect_frequency_anomaly(frequency_log) that:Receives a list of floats called frequency_log.Checks the list to see if there is any spot where two readings in a row both exceed $50.5\text{ Hz}$.If such a back-to-back spike is found anywhere in the list, return True immediately.If you check the entire list and never find two consecutive spikes, return False."""
def detect_frequency_anomaly(frequency_log):
    # --- WRITE YOUR SOLUTION BELOW ---
    for i in  range (1,len(frequency_log)):
      if frequency_log[i] > 50.5 and frequency_log[i - 1]> 50.5:
        return "True"
    return "False"

    # --- WRITE YOUR SOLUTION ABOVE ---


# ==========================================
# 🧪 EVALUATION TEST CASES
# ==========================================

# Test A: Has spikes, but they aren't consecutive (Should return False)
print("Test A:", detect_frequency_anomaly([50.1, 50.6, 49.9, 50.6, 50.2]))

# Test B: Has two consecutive spikes at the end (50.7 and 50.8) -> Should return True
print("Test B:", detect_frequency_anomaly([50.1, 50.2, 50.7, 50.8, 50.1]))
```
