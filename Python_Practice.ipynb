```python
def my_list(list):
  return sorted(list, reverse = True)
data = [1,5,7,29,2,9]
print(f"The list is {my_list(data)}")
even_num = [num for num in data if num %2 == 0 ]
odd_num = [num for num in data if num %2 != 0 ]

def is_prime(num):
  if num == 1:
    return False
  for i in range(2,int(num**0.5)+1):
    if num % i == 0:
      return False
  return True
prime_num = [num for num in data if is_prime(num)]
print(f"Even numbers in the list {even_num}")
print(f"Odd numbers in the list {odd_num}")
print(f"Prime numbers in the list {prime_num}")
```

```python
"""The Challenge: The Grid Telemetry Parser
Write a parametric function named process_market_packet(data_packet) that:

Unpacks the raw tuple using the asterisk (*) operator to isolate the substation name, baseline price, dynamic surges, and stability flag without hardcoded list index numbers.

Finds the maximum hourly price surge inside that dynamic middle group. If no surges occurred, safely default it to 0.0 without breaking.

Packages the output into a namedtuple blueprint called MarketReport with fields: node, peak_surge, and stable_status."""
def process_market_packet(data_packet):
  substation_name,baseline_price,*dynamic_surges,stability_flag = data_packet
  max_price = max(dynamic_surges,default = 0.0)
  from collections import namedtuple
  MarketReport = namedtuple("MarketReport",["node", "peak_surge"  , "stable_status" ])
  return MarketReport(substation_name, max_price, stability_flag)

packet_A = ("Karvenagar_Node", 4.50, 12.80, 15.20, 9.40, True)
packet_B = ("Kothrud_Sub", 5.20, False)

print("Report A:", process_market_packet(packet_A))
print("Report B:", process_market_packet(packet_B))
```
